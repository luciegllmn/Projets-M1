# Importation et manipulation des données 
Nous allons manipuler le jeu de données `data_vin` trouvé à l'[adresse suivante](https://www.kaggle.com/rajyellow46/wine-quality).  

Notre jeu de données comprend 6 497 observations et 13 variables.

Nous allons décrire nos 13 variables :
- `type` est une variable binaire qui sert à savoir si le vin étudié est du vin rouge ou du vin blanc  
- `fixed acidity` : acides utilisés qui ne s'évaporent pas facilement   [unité : g/L]
- `volatile acidity` : quantité d'acide acétique dans le vin, qui, à des niveaux trop élevés, peut donner un goût désagréable de vinaigre  [unité : g/L]
- `citric acid` : acide citrique, qui, en petite quantité peut ajouter de la "fraîcheur" et de la saveur aux vins [unité : g/L] 
- `residual sugar` : la quantité de sucre restant après l'arrêt de la fermentation, il est rare de trouver des vins avec moins de 1 gramme/litre et les vins avec plus de 45 grammes/litre sont considérés comme doux  [unité : g/L]
- `chlorides` : quantité de sel dans le vin   [unité : g/L]
- `free sulfur dioxide` : dioxyde de soufre libre, empêche la croissance microbienne et l'oxydation du vin  [unité : mg/L]
- `total sulfur dioxide` : dioxyde de soufre total ; à faible concentration, le dioxyde de souffre est généralement indétectable dans le vin, mais à des concentrations de dixodyde de souffre libre supérieures à 50 ppm, il devient évident dans le nez et le goût du vin  [unité : mg/L]
- `density` : densité, elle est proche de celle de l'eau en fonction du pourcentage d'alcool et de la teneur en sucre  
- `pH` : décrit le degré d'acidité ou de basicité d'un vin sur une échelle allant de 0 (très acide) à 14 (très basique) ; la plupart des vins se situent entre 3 et 4 sur l'échelle du pH  
- `sulphates` : sulfates : un additif pour le vin qui peut contribuer aux niveaux de dioxyde de soufre gazeux , qui agit comme un antimicrobien et un antioxydant  [unité : g/L]
- `alcohol` : pourcentage d'alcool contenu dans le vin   [unité : %]
- `quality`: qualité du vin (sur une échelle de 10)

## Nous allons maintenant effectuer une régression linéaire afin de savoir quelles propriétés chimiques influent sur la qualité du vin. 

On commence par diviser notre jeu de données `data_vin2` en deux :  un jeu de données d'apprentissage et un de test.   
Nous allons apprendre le modèle à l'aide d'un échantillon de 80% de nos données et on testera le modèle avec les 20% restants.

On souhaite expliquer la variable `quality`($Y_i$) qui peut être modéliser par une loi de Poisson de paramètre $\lambda_i$ car c'est une variable de comptage qui définit une note sur 10 de la qualité du vin.  

Nous étudions donc le modèle suivant :   
$ln(\lambda_i) = \beta_0 + \beta_1 X_{1,i} + ... + \beta_{11} X_{11,i}$  

Avec : 
- $\beta_0$ qui correspond à l'ordonnée à l'origine du modèle
- les $\beta_k$ pour $k\in[1,11]$ qui sont associés à la i_ème variable explicative
- $X_{k,i}$ qui correspondent aux variables quantitatives de notre étude : ce sont les variables explicatives de notre modèle 

```python
#Annexe(22)
```

D'après les résultats obtenu grâce à la fonction `GLM`, on a : 
- dans la colonne `coef` on trouve les estimations de nos 12 coefficients $\beta_0,...\beta_{11}$
- dans la colonne `P>|z|` on trouve les pvaleurs du test de Student où l'on test :   
$H_0 :\ "\beta_k = 0"$    
$H_1 :\ "\beta_k \ne 0"$  

Ce test permet de savoir si nos coefficients sont significatifs ou non, c'est à dire si la variable explicative associée joue un rôle dans la détermination de la qualité d'un vin.  

En prenant un seuil de 5%, on remarque que toutes les p-valeurs des coefficients, à l'exception de celle de `volatile acidity`, `sulphates` et `alcohol`, sont strictement supérieures à $0.05$ : dans ce cas on accepte $H_0$.  
Pour les trois autres pvaleurs des coefficients qui sont strictement inférieures à $0.05$, on rejette $H_0$ et on décide $H_1$.  

En conclusion, lorsque l'on a accepté $H_0$ les coefficients ne sont pas significatifs et donc les variables explicatives associées n'influent pas sur la qualité du vin.   
On peut donc dire que les variables explicatives qui influent sur notre modèle sont : 
- `volatile acidity`
- `sulphates`
- `alcohol`

```python
#Annexe(22)
#Définition de nos variables à étudier
Y = data_vin2['quality']
X = data_vin2[data_vin2.columns[1:12]]

#Décomposition en deux jeux de données, un jeu 'train' et un jeu de test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)

#On ajoute une constante pour effectuer notre régression car GLM fait la régression sans constante par défault
X_train_beta0 = sm.tools.add_constant(X_train)

#Régression linéaire de poisson : 
reg_poiss = sm.GLM(Y_train, X_train_beta0, family=sm.families.Poisson()).fit()
print(reg_poiss.summary())
```

```python
#Annexe(23)
X_test_beta0 = sm.tools.add_constant(X_test)

pred_poiss = reg_poiss.get_prediction(X_test_beta0)
predictions_frame = pred_poiss.summary_frame()

predicted_counts = predictions_frame['mean']
actual_counts = Y_test
fig = plt.figure(figsize = (20,5))
fig.suptitle('Prédiction')
predicted, = plt.plot(X_test.index, predicted_counts, '.', label='Predicted counts')
actual, = plt.plot(X_test.index, actual_counts, '.', label='Actual counts')
plt.legend(handles=[predicted, actual])
plt.show()
```
